```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

Reading Data

```python
df = pd.read_csv('/content/dailyActivity_merged.csv')
df
```

Numbers of users

```python
df.Id.nunique()
```

```python
cols = ['Id',	'ActivityDate',	'TotalSteps', 'VeryActiveMinutes',	'FairlyActiveMinutes',	'LightlyActiveMinutes',	'SedentaryMinutes',	'Calories']
df = df[cols]
df
```

```python
df.rename(columns={'ActivityDate':'Date'}, inplace=True)
df
```

Choosing the important columns

```python
df['Total Minutess'] = df['VeryActiveMinutes'] + df['FairlyActiveMinutes'] + df['LightlyActiveMinutes'] + df['SedentaryMinutes']
df
```

Adding new column called Total Hours

```python
df['Total Hours'] = round(df['Total Minutess'] / 60)
df
```

```python
df.info()
```

```python
df.Date = pd.to_datetime(df['Date'])
df.info()
```

Knowing name of the day

```python
df['Day Of Week'] = df['Date'].dt.day_name()
df
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Day Of Week'], bins = 7, color = 'b', width = .6)
plt.xticks(rotation=90)
plt.grid(True)
plt.title('Day Of Week')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['TotalSteps'], df['Calories'], c = df['Calories'])
plt.xlabel('Total Steps')
plt.ylabel('Calories')
plt.title('Calories vs Total Steps')
median_steps = df['TotalSteps'].median()
median_calories = df['Calories'].median()
plt.axhline(median_calories, color='r', linestyle='--', label='Median Calories')
plt.axvline(median_steps, color='b', linestyle='--', label='Median Steps')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Total Hours'], df['Calories'], c = df['Calories'])
plt.xlabel('Total Hours')
plt.ylabel('Calories')
plt.title('Calories vs Total Hours')
median_Hours = df['Total Hours'].median()
median_calories = df['Calories'].median()
plt.axhline(median_calories, color='r', linestyle='--', label='Median Calories')
plt.axvline(median_Hours, color='b', linestyle='--', label='Median Hours')
plt.legend()
plt.show()
```

```python
VeryActiveMinutes = df['VeryActiveMinutes'].sum()
FairlyActiveMinutes = df['FairlyActiveMinutes'].sum()
LightlyActiveMinutes = df['LightlyActiveMinutes'].sum()
SedentaryMinutes = df['SedentaryMinutes'].sum()
minutes = [VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes]
labels = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary']
plt.pie(minutes, labels=labels, autopct='%1.1f%%', explode=(0, 0, 0, 0.1))
plt.title('Minutes per Category')
plt.show()
```
